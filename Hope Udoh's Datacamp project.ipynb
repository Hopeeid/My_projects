#### 1. What is the average rating by country of origin?

```python
# to see all the columns
chocolate.columns
```

```python
# to find the average rating by using the groupby function
avg_rating = chocolate.groupby('bean_origin', as_index = False)[['rating']].mean().sort_values(by='rating',ascending=False)
avg_rating
```

#### Question 2. How many bars were reviewed for each of those countries?


```python
review_i = chocolate.groupby("bean_origin", as_index = False)[["review"]].count().sort_values(by='review',ascending=False)
review_i
```

#### 3. Create plots to visualize findings for questions 1 and 2.

```python
plt.figure(figsize = (20, 5))
sns.barplot(x = "bean_origin", y = "rating", data = avg_rating[:15])
plt.xticks(rotation = 90);
```

```python
plt.figure(figsize = (20, 5))
sns.barplot(x = "bean_origin", y = "rating", data = avg_rating[48:])
plt.xticks(rotation = 90);
```

- The bean from Tobago has the highest ratings
- The bean from Puerto Rico has the lowest ratings

```python
# to visualize question 2
plt.figure(figsize = (20, 5))
sns.barplot(x = "bean_origin", y = "review", data = review_i)
plt.xticks(rotation = 90);
```

-From this chart it can be said that Venezuela has the most number of review

#### 4. Is the cacao bean's origin an indicator of quality?

```python
# to check if the bean's origin is an indicator of quality, it would be evaluated based on the pattern 
#of rating/bean_origin chart and cocoa_content/rating chart
plt.figure(figsize=(20,5))
sns.lineplot(x='bean_origin', y='rating', data= chocolate)
plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(20,5))
sns.lineplot(x='cocoa_percent', y='rating', data= chocolate)
plt.xticks(rotation=90);
```

- From both charts, it can be said that there is no similar pattern between the charts.-
- it can also be said that cocao bean origin does not indicate quality

#### Question 4
- How does cocoa content relate to rating? What is the average cocoa content for bars with higher ratings (above 3.5)?

```python
# to check the relationship between cocoa content and rating
plt.figure(figsize=(20,5))
sns.barplot(x='cocoa_percent', y='rating', data=chocolate)
plt.xticks(rotation=90);
```

- From the above barplot, it is observed that there is no relationship between Cocoa percent and rating.
- 50% cocoa content has the highest rating above 3.5, and 63% cocoa content has a rating of 3.5
- There is no clear pattern indicating realationship between cocoa percent and rating.
 

```python
#What is the average cocoa content for bars with higher ratings (above 3.5)?
high_rating=chocolate.loc[chocolate.rating>3.5]
aver_cocoa_cont = high_rating.cocoa_percent.mean()
aver_cocoa_cont
```

- The average cocoa content  for bars with higher rating is approximately 71%
